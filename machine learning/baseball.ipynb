# scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x1=sc.fit_transform(x)
```

```python
x1=pd.DataFrame(x1,columns=x.columns)
x1.head()
```

# testing on various models

```python
x1.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor

ln=LinearRegression()
mnb=MultinomialNB()
dtr=DecisionTreeRegressor()
svr=SVR()
knr=KNeighborsRegressor()

rfr=RandomForestRegressor()
abr=AdaBoostRegressor()
gbr=GradientBoostingRegressor()
```

```python
from sklearn import metrics
import math
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
model=[ln,dtr,knr,rfr,gbr,abr]
for m in model:
    m.fit(x1_train,y_train)
    score=m.score(x1_train,y_train)
    pred=m.predict(x1_test)
    r2=r2_score(y_test,pred)
    print('model score of',m,"is :",score)
    print("accuracy score of",m,"is:",r2)
    print("\n")
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
ls=Lasso(alpha=0.001)
rd=Ridge(alpha=0.001)
es=ElasticNet(alpha=0.001)
```

```python
model=[ls,rd,es]
for m in model:
    m.fit(x1_train,y_train)
    score=m.score(x1_train,y_train)
    pred=m.predict(x1_test)
    r2=r2_score(y_test,pred)
    print("for model",m,
         "model score:",score,
         "accuracy score:",r2)
```

```python
from sklearn.model_selection import cross_val_score

model=[ln,dtr,svr,knr,rfr,abr,gbr,ls,rd,es]
for m in model:
    cs=cross_val_score(m,x1,y,cv=5)
    print(cs)
    print("accuracy score of",m,"is",cs.mean())
    print("\n")
```

obs- we will take lasso regression model for prediction

# hypertuning

```python
from sklearn.model_selection import GridSearchCV
alpha={"alpha":[0.0001,0.001,0.01,0.1,1,10]}
gs=GridSearchCV(ls,alpha,cv=5)
gs.fit(x1_train,y_train)
print(gs.best_params_)
```

# final model

```python
ls=Lasso(alpha=1)
ls.fit(x1_train,y_train)
score=ls.score(x1_train,y_train)
pred=ls.predict(x1_test)
r2=r2_score(y_test,pred)
mse=metrics.mean_squared_error(y_test,pred)
mae=metrics.mean_absolute_error(y_test,pred)
rmse=math.sqrt(metrics.mean_squared_error(y_test,pred))
print("model score:",score)
print("accuracy score:",r2)
print("mse=",mse,"mae=",mae,"rmse=",rmse)
```
