```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=.30,random_state=42)
```

# checking on different regressors

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor

lr=LinearRegression()
dtr=DecisionTreeRegressor()
svr=SVR()
knr=KNeighborsRegressor()
rfr=RandomForestRegressor()
abr=AdaBoostRegressor()
gbr=GradientBoostingRegressor()
```

```python
from sklearn import metrics
import math
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
model=[lr,svr,knr,dtr,rfr,gbr,abr]
for m in model:
    m.fit(x1_train,y_train)
    score=m.score(x1_train,y_train)
    pred=m.predict(x1_test)
    r2score=r2_score(y_test,pred)
    print('model score of',m,'is:',score)
    print('accuracy score of',m,'is:',r2score)
    print('\n')
    
```

# checking on cross validation score

```python
from sklearn.model_selection import cross_val_score
model=[lr,svr,knr,dtr,rfr,gbr,abr]
for m in model:
    cs=cross_val_score(m,x1,y,cv=5)
    print(cs)
    print('accuracy score of',m,'is:',cs.mean())
    print("\n")
```

# final model

```python
rfr=RandomForestRegressor()
rfr.fit(x1_train,y_train)
score=rfr.score(x1_train,y_train)
predrfr=rfr.predict(x1_test)
r2score=r2_score(y_test,predrfr)
mse=metrics.mean_squared_error(y_test,predrfr)
mae=metrics.mean_absolute_error(y_test,predrfr)
rmse=math.sqrt(mean_squared_error(y_test,predrfr))
print("model score is:",score)
print('accuracy score of model is',r2score)
print("mean squared error:",mse)
print("mean absolute error:",mae)
print("root mean squared error:",rmse)
print('\n')
```
