# Ans4-i- Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
# Ans 4-a- Top 10 ODI teams in men’s cricket along with the records for matches, points andrating

import requests
import bs4
from bs4 import BeautifulSoup

page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page

soup=BeautifulSoup(page.content)
soup

team=soup.find('span',class_='u-hide-phablet')
team.text

teams=[]
for i in soup.find_all('span',class_='u-hide-phablet'):
    teams.append(i.text)
teams    

teams1=teams[0:10]

teams1

points=soup.find('td',class_="rankings-block__banner--points")
points.text

point=[]
for i in soup.find_all('td',class_='table-body__cell u-center-text'):
    point.append(i.text)
point    

a=point[1::2]

a

a.insert(0,points.text)

a

newpoints=a[0:10]

newpoints

match=soup.find('td',class_='rankings-block__banner--matches')
c=match.text
c

x=point[::2]
x

x.insert(0,c)

x

matches=x[0:10]
matches

rating=soup.find('td',class_='rankings-block__banner--rating u-text-right')
z=rating.text.split('\n')[1]
z

ratings=[]
for i in soup.find_all('td',class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
ratings    

ratings

newrating=ratings[0:9]
newrating

newrating.insert(0,z)

newrating

import pandas as pd
Oditeamranking=pd.DataFrame({'Team':teams1,'Matches':matches,"Points":newpoints,"Rating":newrating})
Oditeamranking

```
