# outlier detection and removal

```python
import scipy
from scipy.stats import zscore
z=np.abs(zscore(d))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
d2=d[(z<3).all(axis=1)]
d2
```

```python
d2.shape
```

```python
x=d2.drop(['label'],axis=1)
x.head()
```

```python
y=d2["label"]
y.head()
```

```python
y.shape
```

# Testing on various models

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
lg=LogisticRegression()
dtc=DecisionTreeClassifier()
Rf=RandomForestClassifier()
Mn=MultinomialNB()
```

```python
model=[lg,dtc,Rf]
for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print("Accuracy score of",m,"is")
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

best accuracy score is of RandomForest Classifier, we will use this model for prediction

```python
pred=Rf.predict(x_test)
```

```python
df= pd.DataFrame({"Actual label":y_test,"Predicted label":pred})
df
```

```python
plt.scatter(df['Actual label'],df['Predicted label'])
```
